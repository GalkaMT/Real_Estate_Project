```python
import pandas as pd 
import re 
import seaborn as sns 
```

```python
df = pd.read_csv('_data.csv', index_col = 'Unnamed: 0')
df.head(3)
```

```python
template = re.compile("^[м][\.] [ЁёА-я]{1,20}\s{0,1}(им.){0,1}\
-{0,1}[0-9]{0,4}[ЁёА-я]{0,20}\s{0,1}[ЁёА-я]{0,20} \([0-9]{1,2} (мин пешком){0,1}(мин на машине){0,1}\)$")
```

```python
mask_small = df['Метро'].str.match(template) != True # для того чтобы True обозначались значения несовпавшие с шаблоном
mask_big = df['Метро'].str.match(template) == True
```

```python
df_small = df[mask_small]
df_big = df[mask_big]
```

```python
df_s = df.shape[0]
df_small_s = df_small.shape[0]
df_big_s = df_big.shape[0]
summary = [(i, df[i].isna().sum()/df_s, df_small[i].isna().sum()/df_small_s, df_big[i].isna().sum()/df_big_s
           ) for i in df.columns.tolist()]
```

```python
summ_pd = pd.DataFrame(summary, columns = ['Column Name', 'Main df', 'Small df', 'Big df'])
summ_pd.set_index('Column Name', inplace = True)
summ_pd # отношение NaN величин к размеру датасета (мотивация - MCAR/MAR/MNAR)
```
